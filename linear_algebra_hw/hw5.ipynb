```python


import numpy as np
import matplotlib.pyplot as plt
A = np.random.randn(20,10)
b = np.random.randn(20)

optimal_x = np.linalg.lstsq(A,b,rcond = None)[0]

def graidient(A,x_k):
  mu = 1/(np.linalg.norm(A))**2
  x_updated = x_k - mu*(A.T@(A@x_k -b))
  return x_updated


x_k = np.zeros_like(A[0])
updated_list = []
updated_list.append(np.linalg.norm(optimal_x ))
for i in range(500):
  updated_vector = graidient(A,x_k)
  updated_list.append(np.linalg.norm(updated_vector - optimal_x))
  x_k = updated_vector


plt.title("convergence")
plt.plot(updated_list)
plt.xlabel("iterations")
plt.ylabel("$||x_k - \hat{x}||$")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = {
    1971: 2250,
    1972: 2500,
    1974: 5000,
    1978: 29000,
    1982: 120000,
    1985: 275000,
    1989: 1180000,
    1993: 3100000,
    1997: 7500000,
    1999: 24000000,
    2000: 42000000,
    2002: 220000000,
    2003: 410000000
}
value = np.zeros(len(data))
names = list(data.keys())
for i in range(len(data)):
  value[i] = data[names[i]]
N = np.log10(value)
A = np.zeros([len(data),2])
for i in range(len(data)):
  A[i,:] = np.array([1,names[i] - 1970])
a,b = np.linalg.lstsq(A,N,rcond = None)[0]

#A. 답
print("a. answer")
print(f"theta_1: ",a)
print(f"theta_2: ",b)
plt.scatter(names,N,color = "red")
plt.plot(range(1970,2004),a + b*np.array([i for i in range(0,2004-1970) ]))
plt.grid()
plt.show()
#B. 답
print("b. answer")
print("prediction: ",10**(a+b*(2015-1970)))
print("real: ",4*10**9)
#C. 답
strong = []
s= 0
L = np.arange(1970,2004,1.5)
for i in L:
  strong.append(data[names[0]]* 2**(s))
  s +=1
weak = []
s= 0
D = np.arange(1970,2004,2)
for i in D:
  weak.append(data[names[0]]* 2**(s))
  s +=1
print("c. answer")
plt.scatter(names,N,color = "red",label="true")
plt.plot(range(1970,2004),a + b*np.array([i for i in range(0,2004-1970) ]),label = "my prediction")
plt.plot(L,np.log10(strong),label="strong(per 1.5years)")
plt.plot(D,np.log10(weak),label="weak(per 2years)")
plt.grid()
plt.legend()
plt.show()
```
