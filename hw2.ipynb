## 1.2 ğ‘Œ ~ ğ‘ğ‘–ğ‘›(ğ‘›, ğ‘) ë¥¼ ìƒˆë¡œ ì •ì˜í•œë‹¤. ğ‘‹ì™€ ğ‘Œì˜ PMF ë¥¼ ê°ê° ê·¸ë ¤ ë³´ì‹œì˜¤. ğ‘› = 10.
ğ‘ = 0.3 ìœ¼ë¡œ ê°€ì •í•œë‹¤

```python
import matplotlib.pyplot as plt
import numpy as np

def Factorial(n:int)->int:
  #Factorial ë°˜í™˜ ì´ë•Œ n! ì„ ë°˜í™˜
  a = 1
  n = int(n)
  if n != 0:
    for i in range(1, n+1):
      a = a*i
  else:
    a = 1
  return a

def Combination(n:int,k:int)->int:
  #nCkë¥¼ ë°˜í™˜
  return Factorial(n) // (Factorial(k)*Factorial(n-k))

def PMF(n,p):
  L = []
  q = 1-p
  for i in range(n+1):
    L.append(Combination(n,i) * p**i * (q)**(n-i))
  return np.array(L)
n = 10
p = 0.3


plt.plot(PMF(n,p),label = "$P_X(x)$")
plt.plot(PMF(n,1-p),label = "$P_Y(y)$")
plt.yticks(np.arange(0, 0.3, 0.05))
plt.grid()
plt.legend()
plt.show()
```

```python
import random
trial = 10000

def tri(p):
  success = 0
  for i in range(10):
    if random.random() <= p:
      success +=1
  return success


def simulation(trial):
  L = np.zeros(11)
  for i in range(trial):
    result = tri(0.3)
    L[result] +=1
  return L
L = simulation(trial)
x = np.arange(0,11,1)
fig, ax1 = plt.subplots()
bar_container = ax1.bar(x, np.array(L))
ax1.set_xlabel("$x$")
ax1.set_ylabel("number of samples",color = "blue")
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
ax2.plot(x, PMF(n,p),color = "red")
ax2.set_ylabel("PMF", color='red')
ax2.tick_params(axis='y', labelcolor='red')
lines, labels = [], []
for ax in [ax1, ax2]:
    line, label = ax.get_legend_handles_labels()
    lines += line
    labels += label
plt.grid(True)
plt.show()

```
