## 1.2 ùëå ~ ùëèùëñùëõ(ùëõ, ùëû) Î•º ÏÉàÎ°ú Ï†ïÏùòÌïúÎã§. ùëãÏôÄ ùëåÏùò PMF Î•º Í∞ÅÍ∞Å Í∑∏Î†§ Î≥¥ÏãúÏò§. ùëõ = 10.
ùëù = 0.3 ÏúºÎ°ú Í∞ÄÏ†ïÌïúÎã§

```python
import matplotlib.pyplot as plt
import numpy as np

def Factorial(n:int)->int:
  #Factorial Î∞òÌôò Ïù¥Îïå n! ÏùÑ Î∞òÌôò
  a = 1
  n = int(n)
  if n != 0:
    for i in range(1, n+1):
      a = a*i
  else:
    a = 1
  return a

def Combination(n:int,k:int)->int:
  #nCkÎ•º Î∞òÌôò
  return Factorial(n) // (Factorial(k)*Factorial(n-k))

def PMF(n,p):
  L = []
  q = 1-p
  for i in range(n+1):
    L.append(Combination(n,i) * p**i * (q)**(n-i))
  return np.array(L)
n = 10
p = 0.3


plt.plot(PMF(n,p),label = "$P_X(x)$")
plt.plot(PMF(n,1-p),label = "$P_Y(y)$")
plt.yticks(np.arange(0, 0.3, 0.05))
plt.grid()
plt.legend()
plt.show()
```

Ïã§Ï†ú ÏÉòÌîåÎßÅÏùÑ ÌÜµÌï¥ ÏúÑÏóêÏÑú Íµ¨Ìïú PMF $ùëÉ_ùëã(ùë•)$ Í∞Ä Í∑∏Î†§ÏßÄÎäîÏßÄ ÌôïÏù∏ÌïúÎã§.
10000 Î≤àÏùò ÏãúÌñâ(trial)ÏùÑ ÌÜµÌï¥ Í∞Å ÌëúÎ≥∏Í∞íÏóê ÏÜçÌïòÎäî ÏÉòÌîå Í∞úÏàòÎ•º Íµ¨ÌïòÍ≥† $ùëÉ_ùëã(ùë•)$ÏôÄ ÎπÑÍµêÌïòÏãúÏò§.

ÏïÑÎûò Í∑∏Î¶ºÏùÑ Ï∞∏Í≥†ÌïúÎã§. (rand Ìï®Ïàò ÌäπÏÑ±ÏúºÎ°ú ÎßâÎåÄ Í∑∏ÎûòÌîÑÎäî Îã§Î•¥Í≤å ÎÇòÏò¨ Ïàò ÏûàÏùå.)

```python
import random
trial = 10000

def tri(p):
  success = 0
  for i in range(10):
    if random.random() <= p:
      success +=1
  return success


def simulation(trial):
  L = np.zeros(11)
  for i in range(trial):
    result = tri(0.3)
    L[result] +=1
  return L
L = simulation(trial)
x = np.arange(0,11,1)
fig, ax1 = plt.subplots()
bar_container = ax1.bar(x, np.array(L))
ax1.set_xlabel("$x$")
ax1.set_ylabel("number of samples",color = "blue")
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
ax2.plot(x, PMF(n,p),color = "red")
ax2.set_ylabel("PMF", color='red')
ax2.tick_params(axis='y', labelcolor='red')
lines, labels = [], []
for ax in [ax1, ax2]:
    line, label = ax.get_legend_handles_labels()
    lines += line
    labels += label
plt.grid(True)
plt.show()

```
